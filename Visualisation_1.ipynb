# Histogram

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) #min=5,max=87
#Sorted array([ 5,  5, 11, 20, 22, 27, 31, 43, 51, 54, 55, 56, 73, 79, 87])
# X = bins range
# Y = how many numbers covered?
#Default bin range = (max-min)/10 = (87-5)/10 = 8.3
#So, bins range = (5-13.3),(13.3-21.6),(21.6-29.9),(29.9-38.2),(38.2-46.5),
#                    (46.5-54.8),(54.8-63.1),(63.1-71.4),(71.4-79.7),(79.7-88)

plt.hist(a) 
plt.title("Histogram")
np.sort(a)
```

```python
plt.hist(a, bins=20) 
plt.title("Histogram")
np.sort(a)
```

# Pi Chart

```python
labels1 = 'Python', 'C++', 'Ruby', 'Java'
#sizes = [215, 130, 245, 210]
sizes = [3,3,2,2]
colors1 = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

#explode1 = (0,0,0,0) #explode list slice
explode1 = (0.4,0,0,0) #explode list slice
#explode1 = (0,0,0.6,0) #explode list slice
#explode1 = (0.2,0.4,0.5,0.6) #explode list slice

#plot
plt.pie(sizes, explode=explode1, labels=labels1, colors=colors1, autopct='%1.1f%%', shadow=False)
plt.axis('equal')
```
