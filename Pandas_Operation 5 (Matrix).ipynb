# Merge Operations

# Merge try to merge column keys

```python
# Same key name

d1 = {'key':['K0','K8','K2','K3'],
     'A':['A0','A1','A2','A3'],
     'B':['B0','B1','B2','B3']}
df_left = pd.DataFrame(d1)

d2 = {'key':['K0','K1','K2','K3'],
     'C':['C0','C1','C2','C3'],
     'D':['D0','D1','D2','D3']}
df_right = pd.DataFrame(d2)
```

```python
df_left
```

```python
df_right
```

When both dataframe have same ke name, at that time set (on=) parameter.

```python
# inner Merge (Chhedgan) 
merge1 = pd.merge(df_left, df_right, how='inner', on='key')
merge1
```

```python
# outer merge (Yogagan)
merge2 = pd.merge(df_left, df_right, how='outer', on='key')
merge2
```

```python
# left merge
merge3 = pd.merge(df_left, df_right, how='left', on='key')
merge3
```

```python
# right merge
merge4 = pd.merge(df_left, df_right, how='right', on='key')
merge4
```

# JOIN Operations

# JOIN try to join index(ROWS) keys

# In join operations, the columns name of both dataframes should be different.
# Otherwise it will give an error

```python
df_left = pd.DataFrame({'A':['A0','A1','A3'], 
                        'B':['B0','B1','B2']}, 
                       index=['K0','K1','K2'])
df_right = pd.DataFrame({'C':['C0','C2','C3'],
                         'D':['D0','D2','D3']},
                       index=['K0','K2','K3'])
```

```python
df_left
```

```python
df_right
```

```python
# inner join, B default how='inner'
df_left.join(df_right)
```

```python
df_right.join(df_left)
```

```python
# Left join
df_left.join(df_right, how='left')
```

```python
df_right.join(df_left, how='right')
```

```python
# Right join
df_left.join(df_right, how='right')
```

```python
df_right.join(df_left, how='left')
```

```python
# Outer join
df_left.join(df_right, how='outer')
```

```python
df_right.join(df_left, how='outer')
```

# apply() used for mapping existing column

```python
# Define function
def testfunc(x):
    if (x>500):
        return (10*np.log10(x))
    else:
        return (float(x)/10)
```

```python
df['col10'] = df['col1']*10
```

```python
df
```

```python
df['FuncApplied'] = df['col2'].apply(lambda x : np.log(x))
```

```python
df
```

```python
df['col20'] = df['col2'].apply(testfunc)
```

```python
df
```

```python
df['col3length'] = df['col3'].apply(len)
df
```

```python
df['squareroot'] = df['col2'].apply(lambda x: np.sqrt(x))
df
```
